```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
food = pd.read_csv('../data/food_ratings - DS003.csv', index_col='student')
food.drop(['age','favorite_color'], axis=1, inplace=True)
food
```

```python
(
food
    .reset_index()
    .melt(id_vars=['student'], value_vars=food.columns)
    .query('value != 0')
    .groupby('variable')
    .agg(mean_rating = ('value', 'mean'))
    .reset_index()
    .sort_values('mean_rating', ascending=False)
    )
```

```python
dense_matrix = (
    food
    .reset_index()
    .melt(id_vars=['student'], value_vars=food.columns,
          var_name='item', value_name='rating')
    .rename(columns={'student':'user'})
    .query('rating != 0')
    )
dense_matrix
```

```python
(
    dense_matrix
    .filter(['user','item','rating'])
    .pivot('user','item','rating')
)
```

```python
from sklearn.metrics.pairwise import cosine_distances
sample=food.copy()
sample_similitude = pd.DataFrame(1 - cosine_distances(sample), columns=sample.index, index=sample.index)
sample_similitude
```

```python
print('brussels_sprouts ratings', sample.brussels_sprouts)
print('Student similitudes to Shikha', sample_similitude.Akbar)
```

```python
weighted_ratings = sum(sample.brussels_sprouts * sample_similitude.Akbar)
weighted_ratings
```
